# convert

```python
# dfpop['State'] = 'Washington'

dfpop['County'].unique()
```

```python
df_comp = pd.read_csv('../data/unemployment.csv')
df_comp.head()
```

```python
df_wcounty = df_comp[df_comp['state'] == 'Washington']
```

```python
# aggregate by County and count the number of records
df_newd = dfpop.groupby('County').agg({'DOL Vehicle ID': 'count'})
```

```python
df_newd = df_newd.reset_index()
df_newd['county'] = df_newd['County'] + ' County'

df_newd['count'] = df_newd['DOL Vehicle ID']
```

```python
df_out_c = df_newd.merge(df_wcounty, on='county')
```

```python
df_out_c
```

```python
df_ouput_county = df_out_c[['id', 'state', 'county', 'DOL Vehicle ID']]
```

```python
df_ouput_county.columns = ['id', 'state', 'county', 'rate']
```

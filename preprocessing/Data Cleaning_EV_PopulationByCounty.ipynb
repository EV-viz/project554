```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Dataset

Dataset landing page: [Electric Vehicle Population Size History By County](https://data.wa.gov/Transportation/Electric-Vehicle-Population-Size-History-By-County/3d5d-sdqb)

```python
# Load data as dataframe
file_path = '../data_raw/Electric_Vehicle_Population_Size_History_By_County.csv'
ev_data = pd.read_csv(file_path)
```

```python
ev_data.head()
```

1. `Date`: The date when the data was recorded.
2. `County`: This is the geographic region of a state that a vehicle's owner is listed to reside within. Vehicles registered in Washington state may be located in other states.
3. `State`: This is the geographic region of the country associated with the record. These addresses may be located in other states.
4. `Vehicle Primary Use`: TThis describes the primary intended use of the vehicle.
5. `Battery Electric Vehicles (BEVs)`: The count of vehicles that are known to be propelled solely by an energy derived from an onboard electric battery.
6. `Plug-In Hybrid Electric Vehicles (PHEVs)`: The count of vehicles that are known to be propelled from energy partially sourced from an onboard electric battery that is capable of being recharged from an external source.
7. `Electric Vehicle (EV) Total`: The sum of Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs).
8. `Non-Electric Vehicle Total`: The count of vehicles that are not electric vehicles.
9. `Total Vehicles`: The All powered vehicles registered in the county. This includes electric vehicles.
10. `Percent Electric Vehicles`: Comparison of electric vehicles versus their non-electric counterparts.

## Distribution of BEVs and PHEVs

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Plot distributions of BEVs and PHEVs
plt.figure(figsize=(14, 6))

# Distribution of BEVsb
plt.subplot(1, 2, 1)
sns.histplot(ev_data['Battery Electric Vehicles (BEVs)'], bins=50, kde=False)
plt.title('Distribution of BEVs')
plt.xlabel('Number of BEVs')
plt.ylabel('Frequency')
plt.yscale('log')  # Log scale due to wide variance

# Distribution of PHEVs
plt.subplot(1, 2, 2)
sns.histplot(ev_data['Plug-In Hybrid Electric Vehicles (PHEVs)'], bins=50, kde=False)
plt.title('Distribution of PHEVs')
plt.xlabel('Number of PHEVs')
plt.ylabel('Frequency')
plt.yscale('log')  # Log scale due to wide variance

plt.tight_layout()
plt.show()
```

The distributions of both Battery Electric Vehicles (BEVs) and Plug-In Hybrid Electric Vehicles (PHEVs) are highly skewed to the right, with most of the data points clustered near the lower end of the scale. This indicates that a large number of counties have few electric vehicles, while a small number of counties have a very high number of electric vehicles.

```python
# Convert 'Date' to datetime format for easier handling
ev_data['Date'] = pd.to_datetime(ev_data['Date'], errors='coerce')

# Group the data by Date and sum up the electric vehicles and total vehicles
ev_time_series = ev_data.groupby('Date').agg({
    'Battery Electric Vehicles (BEVs)': 'sum',
    'Plug-In Hybrid Electric Vehicles (PHEVs)': 'sum',
    'Electric Vehicle (EV) Total': 'sum',
    'Total Vehicles': 'sum'
}).reset_index()

# Calculate the overall percentage of electric vehicles over time
ev_time_series['Percent Electric Vehicles'] = (ev_time_series['Electric Vehicle (EV) Total'] / 
                                               ev_time_series['Total Vehicles']) * 100
```

```python
# Plot the time series of electric vehicles
plt.figure(figsize=(14, 7))

# Plotting total number of EVs over time
plt.plot(ev_time_series['Date'], ev_time_series['Electric Vehicle (EV) Total'], label='Total EVs', color='blue', marker='o')

# Highlighting BEVs and PHEVs in the same plot
plt.plot(ev_time_series['Date'], ev_time_series['Battery Electric Vehicles (BEVs)'], label='BEVs', color='green', linestyle='--')
plt.plot(ev_time_series['Date'], ev_time_series['Plug-In Hybrid Electric Vehicles (PHEVs)'], label='PHEVs', color='red', linestyle='--')

plt.title('Electric Vehicles over Time')
plt.xlabel('Date')
plt.ylabel('Number of Vehicles')
plt.legend()
plt.show()
```

It appears that there is an overall increasing trend in the number of electric vehicles, which suggests that EV adoption may be growing.

```python
# Grouping by states on the number of electric vehicles
state_ev_totals = ev_data.groupby('State').agg({'Electric Vehicle (EV) Total': 'sum'}).reset_index()

# Sorting the states by the total number of electric vehicles
state_ev_totals = state_ev_totals.sort_values(by='Electric Vehicle (EV) Total', ascending=False)

# Plot the distribution of electric vehicles by state
plt.figure(figsize=(14, 10))
sns.barplot(data=state_ev_totals, x='Electric Vehicle (EV) Total', y='State', orient='h')
plt.title('Distribution of Electric Vehicles by State')
plt.xlabel('Total Number of Electric Vehicles')
plt.ylabel('State')
plt.show()
```

# Data Cleaning

```python
# The following steps will be performed:
# 1. Handle missing values in the 'County' and 'State' columns.
# 2. Ensure that the 'Date' column is in the correct datetime format and handle any anomalies.
# 3. Check for any duplicate rows and remove them if necessary.
# 4. Validate the data to ensure that 'Electric Vehicle (EV) Total' is indeed the sum of BEVs and PHEVs.
# 5. Validate that 'Total Vehicles' is the sum of 'Electric Vehicle (EV) Total' and 'Non-Electric Vehicle Total'.
# 6. Check for any anomalies in 'Percent Electric Vehicles' such as values over 100%.

# First, we will drop rows with missing 'County' or 'State' values because without them we cannot accurately attribute the data.
ev_data_cleaned = ev_data.dropna(subset=['County', 'State'])

# Next, check for any anomalies in 'Date' after conversion to datetime.
date_anomalies = ev_data_cleaned[ev_data_cleaned['Date'].isnull()]

# Check for duplicate rows.
duplicates = ev_data_cleaned.duplicated().sum()

# Validate the totals for EVs and overall vehicle counts.
ev_total_mismatch = ev_data_cleaned[ev_data_cleaned['Electric Vehicle (EV) Total'] != ev_data_cleaned['Battery Electric Vehicles (BEVs)'] + ev_data_cleaned['Plug-In Hybrid Electric Vehicles (PHEVs)']]
total_vehicles_mismatch = ev_data_cleaned[ev_data_cleaned['Total Vehicles'] != ev_data_cleaned['Electric Vehicle (EV) Total'] + ev_data_cleaned['Non-Electric Vehicle Total']]

# Check for any anomalies in 'Percent Electric Vehicles'.
percent_ev_anomalies = ev_data_cleaned[ev_data_cleaned['Percent Electric Vehicles'] > 100]

# If there are no anomalies or mismatches, we will have a cleaned dataframe.
# Let's return the findings from the checks.
(date_anomalies.shape[0], duplicates, ev_total_mismatch.shape[0], total_vehicles_mismatch.shape[0], percent_ev_anomalies.shape[0])
```
